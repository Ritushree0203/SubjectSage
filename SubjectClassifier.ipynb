## Web-scraping using BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
cats = ['Biology', 'Mathematics','Physics','Chemistry','History','Geography','Political_science']
```

```python
dfname = []
for cat in cats:
  url = f"https://en.wikipedia.org/wiki/{cat}"
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')
  p_class = soup.find_all('p')
  p_txt = []
  for i in range(len(p_class)):
    para = p_class[i].get_text()
    sent = para.split('.')
    for i in range(len(sent)):
      p_txt.append(sent[i])
  df = pd.DataFrame(p_txt, columns = ['Text'])
  df['Category'] = f"{cat}"
  dfname.append(df)
```

```python
Master_df = pd.DataFrame()
for i in range(len(dfname)):
  Master_df = pd.concat([Master_df,dfname[i]],axis=0)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
bio_df = df[df['Category'] == 'Biology']
bio_vect = CountVectorizer()
bio_vect.fit(bio_df['Text'])
bio_vocab = bio_vect.get_feature_names_out()
bio_vocab = list(bio_vocab)

math_df = df[df['Category'] == 'Mathematics']
math_vect = CountVectorizer()
math_vect = math_vect.fit(math_df['Text'])
math_vocab = math_vect.get_feature_names_out()
math_vocab = list(math_vocab)

phy_df = df[df['Category'] == 'Physics']
phy_vect = CountVectorizer()
phy_vect = phy_vect.fit(phy_df['Text'])
phy_vocab = phy_vect.get_feature_names_out()
phy_vocab = list(phy_vocab)

chem_df = df[df['Category'] == 'Chemistry']
chem_vect = CountVectorizer()
chem_vect = chem_vect.fit(chem_df['Text'])
chem_vocab = chem_vect.get_feature_names_out()
chem_vocab = list(chem_vocab)

hist_df = df[df['Category'] == 'History']
hist_vect = CountVectorizer()
hist_vect = hist_vect.fit(hist_df['Text'])
hist_vocab = hist_vect.get_feature_names_out()
hist_vocab = list(hist_vocab)

geo_df = df[df['Category'] == 'Geography']
geo_vect = CountVectorizer()
geo_vect = geo_vect.fit(geo_df['Text'])
geo_vocab = geo_vect.get_feature_names_out()
geo_vocab = list(geo_vocab)

ps_df = df[df['Category'] == 'Political_science']
ps_vect = CountVectorizer()
ps_vect = ps_vect.fit(ps_df['Text'])
ps_vocab = ps_vect.get_feature_names_out()
ps_vocab = list(ps_vocab)
```

```python
for index, row in df.iterrows():
  words = row['Text'].split()
  new_words = []
  for word in words:
    if(word in bio_vocab and word in chem_vocab and word in hist_vocab and word in geo_vocab and word in ps_vocab and word in math_vocab and word in phy_vocab):
      words.remove(word)
      df.at[index, 'Text'] = ' '.join(words)
```

```python
df['Category'].value_counts()
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def tokenize(text):
  return [lemmatizer.lemmatize(token) for token in word_tokenize(text) if token.isalpha()]
english_stopwords = stopwords.words('english')
vectorizer = TfidfVectorizer(tokenizer=tokenize, stop_words=english_stopwords, ngram_range=(1,2))
vectorizer.fit(df['Text'])
```

```python
vectorizer.get_feature_names_out()[:50]
```

```python
vect_inputs = vectorizer.transform(df['Text'])
```

```python
df_label_encoder = df.copy()
```

### Label-encoding categorical column "Category"

```python
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(df_label_encoder['Category'])
df_label_encoder['Category'] = y_encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(vect_inputs, df_label_encoder['Category'], test_size=0.3, random_state=42)
```

## Model Training
*   Logistic Regression (Multinomial classification)
*   RandomForest Classifier
*   Support Vector Classifier
*   XGBClassifier
*   GridSearchCV

### Logistic regression

```python
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=5000)  # Use 'lbfgs' solver for multi-class problems
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

```python
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))

plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
